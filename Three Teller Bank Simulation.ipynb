```python
import numpy as np

def hizmet_suresi(vezne_numarasi):
    # Her veznenin hizmet süresi üstel dağılıma uysun
    if (vezne_numarasi == 1):
        return np.random.exponential(scale=3)
    elif (vezne_numarasi == 2):
        return np.random.exponential(scale=5)
    elif (vezne_numarasi == 3):
        return np.random.exponential(scale=7)

def ogrenci_hizmet(ogrenci_akisi, hizmet_ayar):
    # Veznelerin başlangıç durumu
    vezne_durumu = [0, 0, 0]
    kuyruk = [[] for _ in range(3)] # Her vezne için ayrı kuyruk
    sonraki_ogrenci_zamani = 0

    for ogrenci in range(1, ogrenci_akisi + 1):
        # Hangi vezneye gideceği belirleniyor (en kısa sıradaki vezne seçiliyor)
        vezne_numarasi = np.argmin(vezne_durumu) + 1

        # Öğrencinin kuyruğa bekleyeceği süre
        bekleme_suresi = max(0, vezne_durumu[vezne_numarasi - 1] - sum(kuyruk[vezne_numarasi - 1]))

        # Öğrencinin hizmet süresi
        hizmet_sure = hizmet_suresi(vezne_numarasi)

        # Vezne durumunu ve kuyruğu güncelleme
        vezne_durumu[vezne_numarasi - 1] += hizmet_sure
        kuyruk[vezne_numarasi - 1].append(bekleme_suresi)

        print(f"Öğrenci {ogrenci}: Vezne {vezne_numarasi} tarafından {hizmet_sure:.2f} dakika süreyle hizmet aldı.")
        print(f"Bekleme Süresi: {bekleme_suresi:.2f} dakika")

        # Sonraki öğrencinin zamanını belirleme
        sonraki_ogrenci_zamani += np.random.exponential(scale=hizmet_ayar)

    # İstatistikleri gösterme
    toplam_hizmet_suresi = sum(vezne_durumu)
    ortalama_bekleme_suresi = np.mean([item for sublist in kuyruk for item in sublist])

    print("\nİstatistikler:")
    for i in range(3):
        print(f"Vezne {i + 1} Meşgul Süre: {vezne_durumu[i]:.2f} dakika")
        print(f"Vezne {i + 1} Boş Süre: {toplam_hizmet_suresi - vezne_durumu[i]:.2f} dakika")
        print(f"Vezne {i + 1} Ortalama Bekleme Süresi: {np.mean(kuyruk[i]):.2f} dakika\n")

    print(f"Toplam Hizmet Süreleri: {toplam_hizmet_suresi:.2f} dakika")
    print(f"Ortalama Bekleme Süresi: {ortalama_bekleme_suresi:.2f} dakika")

# Saatte 60 öğrenci akışını simüle et, hizmet sürelerini ayarla 
ogrenci_akisi = 60
hizmet_ayar = 60 / ogrenci_akisi # Toplam süreyi 1 saat olarak ayarla

# Fonksiyonu çağırma
ogrenci_hizmet(ogrenci_akisi, hizmet_ayar)
```
